```python
# For Sepeartion of Variable
X= df.drop(columns=['Salary'])
y= df['Salary']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
pd.DataFrame({"y_test:":y_test,"y_pred:":y_pred,"MAE:":y_test-y_pred})
```

```python
#import metrics
import numpy as np
from sklearn.metrics import mean_squared_error,r2_score
print("mean squraed erro:",mean_squared_error(y_test,y_pred))
print("r2_score:",r2_score(y_test,y_pred))
print("root mean squraed erro:",np.sqrt(mean_squared_error(y_test,y_pred)))
```
