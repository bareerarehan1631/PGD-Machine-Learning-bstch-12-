```python
# STEP 1 DATA UNDERSTANDING AND LOADING
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# step DATA PROCESSING
```

```python
df["Gender"].mode() #fill the mssing column gender
```

```python
df["Gender"]=df["Gender"].fillna(df["Gender"].mode()[0])
df
```

```python
df.isnull().sum()
```

```python
df["Age"]=df["Age"].fillna(df["Age"].mean()) #fill the missing column age
df.isnull().sum()
```

```python
df["City"]=df["City"].fillna(df["City"].mode()[0]) # fill the missing column city
```

```python
df.isnull().sum()
```

```python
df["Sleep Duration"]=df["Sleep Duration"].fillna(df["Sleep Duration"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["Degree"]=df["Degree"].fillna(df["Degree"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["Work/Study Hours"]=df["Work/Study Hours"].fillna(df["Work/Study Hours"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["Financial Stress"]=df["Financial Stress"].fillna(df["Financial Stress"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["Family History of Mental Illness"]=df["Family History of Mental Illness"].fillna(df["Family History of Mental Illness"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df= pd.get_dummies(df,drop_first=True)
df.head()
```

```python
df
```

```python
X=df.drop(columns=["Depression","id"])
```

```python
y=df["Depression"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,   # 30% test data
    train_size=0.70,  # 70% training data (optional)
    random_state=42
)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
model
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
pd.DataFrame({"y_test:":y_test,"y_pred:":y_pred})
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print("accuracy_score:",accuracy_score(y_test,y_pred))
```

```python
print("Confusion_matrix:",confusion_matrix(y_test,y_pred))
```

```python
print("classification report:",classification_report(y_test,y_pred))
```
